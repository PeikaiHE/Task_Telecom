### Can we determine the extraction epoch e* (Sec. 3.3) by looking at the output of the Softmax layer? 

And the answer is yes, we can determine the extraction epoch e* by looking at the output of the Softmax layer.

Because the output of the Softmax layer is a probability distribution over the classes, then we can use this kink of information to determine the distance between the Centroids. For example, in this paper, for the t-th class, we can calculate the average of the Softmax outputs of all samples that are correctly classified as the t-th class:
$$C_{e,t}=\frac{1}{|D_{t}^{||,e}|}\sum_{i\in D_{t}^{||,e}}a_{e,i}$$

Then we calculate the distance between the Centroids: 
$$d^{*}(\boldsymbol{a}_{e,i})=\left\{\begin{array}{ll}0&\mathrm{if~}y_{e,i}=\hat{y}_{i}\\2\cdot\frac{\|\boldsymbol{a}_{e,i}\boldsymbol{-H}_{{e,\boldsymbol{C}_{{y_{e,i}}}}\boldsymbol{C}_{{e,\hat{y}_{i}}}}\|_{2}}{\|\boldsymbol{C}_{{y_{e,i}}}\|_{2}+\|\boldsymbol{C}_{{e,\hat{y}_{i}}}\|_{2}}&\mathrm{if~}y_{e,i}\neq\hat{y}_{i},\end{array}\right.$$

And then we can find the $e^{*}$ epoch by looking at the output of the Softmax layer:
$$e^*=\underset{e}{\mathrm{argmax}}\frac{\sum_{i}d^*(\boldsymbol{a}_{e,i})}{\|\mathcal{D}_{e}^\perp\|_0}$$
